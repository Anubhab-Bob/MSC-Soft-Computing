```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rawdata = pd.read_csv("heart.csv")
rawdata.head()
```

```python
rawdata.describe()
```

```python
nrow, ncol = rawdata.shape
print(nrow,ncol)
```

```python
# Checking the datatypes of the attributes
rawdata.dtypes
```

```python
# Get names of attributes
colNames = rawdata.columns.to_list()
print(colNames)
```

```python
# Getting the Covariance matrix
rawdata.cov()
```

```python
# Getting the correlation matrix
rawdata.corr()
```

```python
# Histogram Draw and save to file
# Adjust the bar widths with the rwidth parameter and size of each histogram with figsize
rawdata.hist(rwidth=0.8, figsize=(20, 15), color = 'orange')
plt.tight_layout()
plt.savefig("histo.jpeg")
```

```python
# Save figures
plt.savefig('histo.jpeg')
```

```python
# Save transparent figures
plt.savefig('hito1.jpeg', transparent=True)
```
