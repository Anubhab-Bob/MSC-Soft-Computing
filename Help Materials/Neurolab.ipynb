**Single-layer perceptron**	
*Function:* newp	
*No. of layers:* 1	
*support train function:* train_delta	
*support error function:* SSE

```python
# neurolab.net.newp(minmax,cn,transf)
# minmax: list of list, the outer list is the number of input neurons,inner lists must contain 2 elements - min and max
# cn: int, number of output neurons
# transf: activation function

import neurolab as nl

input = [[0, 0], [0, 1], [1, 0], [1, 1]]
target = [[0], [0], [0], [1]]

# Create net with 2 inputs and 1 neuron
net = nl.net.newp([[0, 1],[0, 1]], 1)

# Delta Rule can be understood by looking it as training an unthresholded perceptron which is trained using gradient descent . The linear combination of weights and the inputs associated with them acts as an input to activation function.
# O(x) = w.x
# Train with delta rule
error = net.train(input, target, epochs=100, show=10, lr=0.1)

# Plot the result
import pylab as pl
pl.plot(error)
pl.xlabel('Epoch number')
pl.ylabel('Train error')
pl.grid()
pl.show()
```
