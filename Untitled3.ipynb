```python
# Realization of a three inputs bipolar NOR gate using perceptron

"""Importing python library."""
import neurolab as nl
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Create a variable named data that is a list that contains the eight possible inputs to an NOR gate."""

data = np.array([[1,0,1,1]])
print(data)
labels = np.array([[0,1]])
#data = data.reshape(len(data),1)
#labels = labels.reshape(len(labels),1)
data.shape[1]
```

```python
# Syntax for 3-D projection
fig = plt.figure()
ax = plt.axes(projection="3d")
for inp,target in zip(data,labels):
        plt.plot(inp[0],inp[1],inp[2],'ro' if (target == 1.0) else 'bo')
plt.legend(['1','-1'])
plt.title("3 input NOR gate")  
plt.show()
```

```python
net = nl.net.newff([[0,1],[0,1],[0,1],[0,1]],[3,3,2])
net.ci
```

```python
error = net.train(data, labels, epochs=500, show=100, goal=0.02)
```

```python
# Simulate network
out = net.sim(data)
print(out)
```

```python
# Plot result
import pylab as pl
pl.subplot(211)
pl.plot(error)
pl.xlabel('Epoch number')
pl.ylabel('error (default SSE)')

x2 = np.linspace(-6.0,6.0,150)
y2 = net.sim(data).reshape(2)

y3 = out.reshape(len(labels[0]))

pl.subplot(212)
pl.plot(data[0], y2, '-',data , labels, '.', data, y3, 'p')
pl.legend(['train target', 'net output'])
pl.show()
```
