```python
import numpy as np
import neurolab as nl
import numpy.random as rand
import matplotlib.pyplot as plt
import pylab as pl
```

```python
# random predefined ceters for 4 clusters
centr = np.array([[0.2, 0.2], [0.4, 0.4], [0.7, 0.3], [0.2, 0.5]])
rand_norm = 0.05 * rand.randn(100, 4, 2)
print(rand_norm)
```

```python
# creating training data
inp = np.array([centr + r for r in rand_norm])
inp.shape = (100 * 4, 2)
rand.shuffle(inp) 
print(inp)
```

```python
inp.shape[0]
centr.shape[0]
```

```python
# Create net with 2 inputs and 50 neurons
net = nl.net.newc([[0.0, 1.0],[0.0, 1.0]], 50)
# train with rule: Winner Take All algoritm (CWTA)
error = nl.train.train_wta(net,inp, epochs=100, goal=1, show=50)
```

```python
# Plot results:
pl.title('Classification Problem')
pl.subplot(211)
pl.plot(error)
pl.xlabel('Epoch number')
pl.ylabel('error (default MAE)')
w = net.layers[0].np['w']

pl.subplot(212)
pl.plot(inp[:,0], inp[:,1], '.', \
        centr[:,0], centr[:, 1] , 'yv', \
        w[:,0], w[:,1], 'p')
pl.legend(['train samples', 'real centers', 'train centers'])
pl.show()
```

```python
w = net.layers[0].np['w']
fig2, ax2 = plt.subplots()
ax2.set_title('Trained Clusters')
for j in range(len(centr)):
    x = centr + net.errorf(centr,w[j])
    ax2.plot(inp[:,0], inp[:,1], 'b.', \
             centr[j][0],centr[j][1], 'yv', \
             x[0], x[1], 'p')
ax2.legend(['train samples', 'real centers', 'train centers'])
#ax2.legend()

print("Plotting on the trained model with the original data (press enter to continue)..")
plt.show()
```
